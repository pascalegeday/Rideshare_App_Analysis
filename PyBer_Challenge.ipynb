# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_by_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_by_type
```

```python
# 2. Get the total drivers for each city type
total_drivers_by_type = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_by_type
```

```python
#  3. Get the total amount of fares for each city type
total_fares_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_by_type
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_per_ride = total_fares_by_type/total_rides_by_type
avg_fare_per_ride.head()
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_driver = total_fares_by_type/total_drivers_by_type
avg_fare_per_driver.head()
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df= pd.DataFrame({"Total Rides":total_rides_by_type, 
                      "Total Drivers":total_drivers_by_type,
                     "Total Fares": total_fares_by_type,
                     "Average Fare per Ride": avg_fare_per_ride,
                     "Average Fare per Driver": avg_fare_per_driver})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df= pd.DataFrame({"Total Rides":total_rides_by_type.map("{:,.0f}".format), 
                      "Total Drivers":total_drivers_by_type.map("{:,.0f}".format),
                     "Total Fares": total_fares_by_type.map("${:,.2f}".format),
                     "Average Fare per Ride": avg_fare_per_ride.map("${:,.2f}".format),
                     "Average Fare per Driver": avg_fare_per_driver.map("${:,.2f}".format)})
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
#Read the merged DataFrame
pyber_data_df.head()
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
total_fare_per_date_type_df = pyber_data_df.groupby(["type","date"]).sum()[["fare"]]
total_fare_per_date_type_df
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
total_fare_per_date_type_df = total_fare_per_date_type_df.reset_index()
total_fare_per_date_type_df.head()
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
total_fare_per_date_type_df = total_fare_per_date_type_df.pivot(index="date", columns="type", values="fare")
total_fare_per_date_type_df
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
fare_date_range_df = total_fare_per_date_type_df.loc["2019-01-01":"2019-04-29"]
fare_date_range_df                                          
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
fare_date_range_df.index = pd.to_datetime(fare_date_range_df.index)
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
fare_date_range_df.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
fare_sum_per_week_df = fare_date_range_df.resample("W").sum()
fare_sum_per_week_df.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
fare_sum_per_week_df.plot(figsize=(17,6))
plt.title("Total Fare by City Type", fontsize=20)
plt.ylabel("Fare ($USD)", fontsize=12)
plt.xlabel(None)
lgnd = plt.legend(fontsize="11", mode="Expanded",loc="center", title="Type")
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(11)
plt.grid(True)
#save image
plt.savefig("Pyber_fare_summary.png", bbox_inches = 'tight')
```
